```python
# problem 7

from sympy import symbols, Matrix, simplify

# Define symbolic variables
m11, m12, m13, m21, m22, m23, m31, m32, m33 = symbols('m11 m12 m13 m21 m22 m23 m31 m32 m33')

# Create the 3x3 matrix
M = Matrix([
    [m11, m21, m31],
    [m12, m22, m32],
    [m13, m23, m33]
])

# Method 1: Using the .inv() method
M_inverse_1 = M.inv()

# Method 2: Using the **-1 operator
# M_inverse_2 = M**-1

# Print the results
print("Original Matrix (M):")
print(M)
print("\nInverse Matrix (M_inverse_1):")
print(simplify(M_inverse_1))

m11 = 1
m12 = 1
m13 = 1
m21 = 0
m22 = 1
m23 = 0
m31 = 0
m32 = 0
m33 = 1

M = Matrix([
    [m11, m21, m31],
    [m12, m22, m32],
    [m13, m23, m33]
])

M_inverse_1 = M.inv()
# Print the results
print("\nOriginal Matrix (M):")
print(M)
print("\nInverse Matrix (M_inverse_1):")
print(simplify(M_inverse_1))
```

```python
# problem 15

# code from google search AI

import itertools
from math import factorial

def permute_and_save(input_list, filename="permutations_6.txt"):
    """
    Generates all permutations of elements in input_list and saves them to a text file.

    Args:
        input_list: The list or other iterable to permute.
        filename: The name of the output text file.
    """
    # Generate the permutations. itertools.permutations returns an iterator of tuples.
    permutations_iterator = itertools.permutations(input_list)

    # Open the file in write mode ('w'). 
    # The 'with' statement automatically handles file closing.
    with open(filename, 'w') as f:
        # Iterate over each permutation as it is generated (lazy evaluation)
        n_no_fixed_point = 0
        for perm in permutations_iterator:
            # Convert the tuple permutation to a string, then add a newline character
            # The elements are converted to strings first to ensure compatibility
            no_fixed_point = all(x != y for x, y in zip(perm, input_list))
            if no_fixed_point:
                line = "".join(map(str, perm)) + ", no fp\n"
                n_no_fixed_point += 1
            else:
                line = "".join(map(str, perm)) + "\n"
            # Write the line to the file
            f.write(line)

    print(f"All permutations have been successfully saved to {filename}.")
    print(f"Number of permutations with no fixed point: {n_no_fixed_point}")
    print(f"Fraction of permutations with no fixed point: {n_no_fixed_point / factorial(len(input_list)):.6f}")

# --- Example Usage ---
my_list = [1, 2, 3, 4, 5, 6]
permute_and_save(my_list)

# The file "permutations.txt" will contain:
```
